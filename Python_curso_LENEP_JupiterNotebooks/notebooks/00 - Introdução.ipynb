SciPy é um ecossistema baseado em Python de softwares matemáticos, ciência e engenharia.
  
Suas principais bibliotecas são:
- _NumPy_
- _MatPlotlib_
- SciPy library

Estaremos estudando numpy, matplotlib e pandas

## NumPy


  
- É usada para realizar cálculos em Arrays Multidimensionais. 
- Fornece um grande conjunto de funções e operações de biblioteca que ajudam os programadores a executar facilmente cálculos numéricos
- Principais aplicações:
    - Modelos de Machine Learning
    - Processamento de Imagem e Computação Gráfica
    - Tarefas Matemáticas

## Matplotlib



- Biblioteca de plotagem 2D 
- Pode gerar gráficos, histogramas, espectros de potência, gráficos de barras, gráficos de dispersão, etc.

## Pandas



- manipulação e análise de dados
- oferece estruturas de dados e operações para manipular tabelas numéricas e séries temporais

## Anaconda
- É como se fosse um sistema operacional do Python/R
- É uma distribuição livre das linguagens de programação Python e R para computação científica 
- Simplifica o gerenciamento de pacotes e desenvolvimento
- Inclui mais de 1400 pacotes populares de ciência de dados



## Spider



- é um ambiente de desenvolvimento integrado (IDE) para programação científica na linguagem Python. 
- Se integra com o SciPy
- Trabalha de forma semelhante ao MatLab
