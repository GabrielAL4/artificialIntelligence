## Funções numpy

```python
import numpy as np

print(np.tan(np.pi/2))

print(np.sin(np.array((0., 30., 45., 60., 90.)) * np.pi / 180. ))
```

```python
a = np.linspace(np.pi*-1, np.pi, 9)
print(a)
print(np.sinh(a))
```

```python
print(np.trunc(3.821))
print(np.around(3.821))
print(np.floor(3.821))
print(np.ceil(3.821))
```

```python
x = np.array([2, 3, 4, 5])
print(np.prod(x))
print(np.sum(x))
```

```python
print(np.exp(2))
print(np.expm1(2))
print(np.log(2))
print(np.log10(2))
print(np.log2(2))
```

```python
print(np.signbit(-11))
print(np.frexp(-1))
```

```python
print(np.angle(1+1j, deg=True))
print(np.real(2-5j))
print(np.conj(2-5j))
```

```python
print("2^3 = ",np.power(2,3))
print("Resto da divisão de 2 por 3: ",np.mod(6,3))
```

```python
#convolução em cada ponto de sobreposição, com uma forma de saída de (N + M-1,). 
#Nos pontos finais da convolução, os sinais não se sobrepõem completamente 
#e os efeitos de contorno podem ser vistos.
print(np.convolve([1, 2, 3], [0, 1, 0.5]))

#O modo "same" retorna a saída de tamanho máximo (M, N). 
#Efeitos de limite ainda são visíveis.
print(np.convolve([1,2,3],[0,1,0.5], 'same'))

#Modo "valid" retorna saída de comprimento max (M, N) - min (M, N) + 1. 
#O produto de convolução é dado apenas para pontos onde os sinais se sobrepõem completamente.
#Valores fora do limite do sinal não têm efeito.
print(np.convolve([1,2,3],[0,1,0.5], 'valid'))

#A função degrau Heaviside
print(np.heaviside([-1.5, 0, 2.0], 0.5))
```
