## Configurando os Eixos e a Grade

Quando um grfico  desenhado, o comando plt.plot(...) automaticamente determina que a escala de cada coordenada varie do menor valor at o maior valor. 
  
Entretanto, muitas vezes deseja-se alterar esses valores, por exemplo deseja-se que a escala de x varie de zero at o valor mximo, e no do valor mnimo ao mximo. 
  
Para especificar novos valores usa-se o comando plt.axis(...). Esse comando possui um argumento, que  uma tupla.
Essa tupla deve conter quatro valores referentes ao x mnimo, x mximo, y mnimo e y mximo que se deseja usar.
  
Por exemplo, para especificar que o eixo x varia de 0 a 1 enquanto o eixo y varia de 0 a 10, a instruo :
  
plt.axis((0,1,0,10))
  
Os parntesis duplos so necessrios porque o comando plt.axis(...) exige uma tupla (ou lista) como argumento.
  
Alternativamente, para configurar os limites dos eixos x e y separadamente usam-se os comandos plt.xlim(...) e plt.ylim(...) respectivamente, com os limites superior e inferior como argumentos. 
  
Assim o exemplo anterior  equivalente a:
  
plt.xlim(0,1)
  
plt.ylim(0,1)
  
Para colocar o desenho de uma grade ao fundo de um grfico, tem-se a comando plt.grid(True). Esse comando requer um nico argumento. Se for True a grade  desenhada, se for False a grade no  desenhada.

```python
x = np.linspace(0,1,num=100)
y = x
z = [t**2 for t in x]
w = [t**3 for t in x]
plt.plot(x,y,'r--',x,z,'bs',x,w,'g^')

#Eixo x varia de 0 a 1 e o eixo y varia de 0 a 10
plt.axis((0,1,0,10))

#Exibe a grade
plt.grid(True)
```

```python
x = np.linspace(0,1,num=100)
y = x
z = [t**2 for t in x]
w = [t**3 for t in x]
plt.plot(x,y,'r--',x,z,'bs',x,w,'g^')

#Eixo x e y variam de 0 a 1
plt.xlim(0,1)
plt.ylim(0,1)

#Exibe a grade
plt.grid(True)
```

### Trabalhando com vrios grficos e janelas

O mdulo matplotlib.pyplot usa o conceito de janela corrente e grfico corrente. 
  
Todos os comandos de desenho so aplicados  janela e ao grfico corrente. A funo plt.figure(...)  o comando que cria, e seleciona, uma janela no terminal onde o grfico ser mostrado. 

Nesse exemplo, e nos anteriores, esse comando  opcional pois se omitido ser considerado como plt.figure(1) quando o grfico for criado pela primeira vez (pelo comando plt.plot(...)).. 
  
Dentro de uma mesma janela, pode haver mais de um grfico. Para criar os diferentes grficos dentro da janela usa-se o comando plt.subplot(...). O argumento do comando plt.subplot(...)  formado por trs valores que representam: Nmero de Linhas, Nmero de Colunas, e Nmero do Grfico.

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.linspace(5,10,num=100)
yt1 = f(t1)
t2 = np.linspace(5,25,num=100)
yt2 = f(t2)

plt.figure(1) # Cria a janela
plt.subplot(2,1,1) # Cria a rea do primeiro grfico
plt.plot(t1, yt1, 'bo', t2, yt2, 'k') # Desenha o grfico

plt.subplot(2,1,2) # Cria a rea do segundo grfico
ycos = np.cos(2*np.pi*t2)

plt.plot(t2, ycos, 'r--') # Desenha o segundo grfico
```

O exemplo abaixo cria duas janelas. 
  
Uma das janelas possui dois grficos e a outra apenas um. 
  
O exemplo mostra como se pode navegar entre janelas e figuras usando plt.figure(...) e plt.subplot(...).

```python
plt.figure(1) # a primeira janela
plt.subplot(2,1,1) # o primeiro grfico na primeira janela

plt.plot((1,2,3))
plt.subplot(2,1,2) # o segundo grfico na primeira janela
plt.plot((4,5,6))

plt.figure(2) # uma segunda janela
plt.plot((4,5,6)) # cria o grfico em subplot(1,1,1) por padro

plt.figure(1) # torna a janela 1 a janela corrente; subplot(2,1,2) ainda  o grfico corrente
plt.subplot(2,1,1) # faz subplot(2,1,1) na janela 1 o grfico corrente
plt.title(" fcil como 1,2,3") # Ttulo do grfico 2,1,1
```

### Usando smbolos matemticos em um texto.
  
Para a escrita de frmulas, aceitam-se as marcaes de frmulas do TeX. Por exemplo para escrever a expresso
  
$\sigma_i=23$' 
  
  em um ttulo pode-se escrever a expresso em
TeX envolto em cifres (como no TeX). Alem disso  necessrio usar o modificador r antes da
string para que a barra invertida no seja confundida com um caractere de escape:
  
plt.title(r  $ \ sigma  i = 23 $)
   
No  necessrio ter o TeX instalado no computador para que as expresses funcionem. 
  
Entretanto se usar marcaes especficas do LaTeX  necessrio que o mesmo esteja instalado.
  
O exemplo abaixo ilustra o uso das funes de posicionamento de texto, incluindo o uso de smbolos matemticos.

```python
from random import random

mu, sigma = 100, 15

x = list() 
for i in range(10000):
    u = random()-0.5
    
    if u >=0 :
        x.append(mu + sigma*np.sqrt(u) )
    else:
        x.append(mu - sigma*np.sqrt(-u))

# Desenha o histograma dos dados
plt.hist(x, 50, density=1, facecolor='g')

plt.xlabel('Resultado')
plt.ylabel('Probabilidade')
plt.title('Histograma do Teste')
plt.text(95, .08, r'$\mu=100,\ \sigma=15$')
plt.axis([85, 115, 0, 0.1])
plt.grid(True)
```

## Destacando pontos
  
Para destacar elementos importantes em um grfico usa-se o comando plt.annotate(...).   
Esse comando marca um ponto no grfico, e coloca um texto (que pode ser diferente do ponto marcado). Opcionalmente desenha uma seta partindo do texto at o ponto marcado.   
Os argumentos da funo so: a string do texto a ser mostrado, uma tupla xy indicando o ponto marcado, uma tupla xytext com as coordenadas da posio do texto.  
Para que a seta aparea  necessrio criar um dicionrio para o argumento arrowprops que contenha a chave facecolor, que d a cor da seta. Esteticamente fica melhor se a seta ficar ligeiramente mais curta. Consegue-se esse efeito colocando a chave shrink no dicionrio com valor de 0.05.


```python
plt.subplot(111)
t = np.linspace(0,5,num=500 )
s = np.cos(2*np.pi*t)
plt.plot(t, s, lw=2)

plt.annotate('mximo local', xy=(2, 1), xytext=(3, 1.5), arrowprops=dict(facecolor='black',shrink=0.05))
plt.ylim(-2,2) 
```
