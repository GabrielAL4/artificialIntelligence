Nesta aplicação prática será lido um arquivo de dados de poços (poco1_adaptado.xls) e serão gerados 3 gráficos diferentes.

Só que nessa atividade, será criada uma função para plotar os gráficos, reduzindo o tamanho do código
A partir desses dados, serão exibidos alguns gráficos, com alguns dados como estão na planilha, outros com os dados alterados ou com variáveis geradas a partir desses dados

```python
#Primeira parte - importar as bibliotecas necessárias
#Neste exercício serão importadas as bibliotecas do pandas para manipulação do arquivo e 
#do matplotlib para plotagem dos gráficos

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Criando a função plot_figure que recebe os dados de x e de y, o título, título do eixo de y
#E os limites dos eixos
def plot_figure(x,y,titulo,titulo_y,xi,xf,yi,yf):
    #x- dados eixo x
    #y- dado eixo y
    #titulo- título do gráfico
    plt.figure()
    plt.plot(x,y,'o',lineWidth=2)
    plt.grid()
    plt.box()
    plt.title(titulo)
    plt.xlabel('\phi (%)')
    plt.ylabel(titulo_y)
    plt.axis([ xi,xf,yi,yf]);
    plt.draw
```

```python
#Abre o primeiro arquivo xls
arqdt = pd.read_excel("poco1_adaptado.xls")
```

```python
#Na ordem da planilha... Capturando dados das colunas necessárias para essa aplicação.


vp=arqdt['VP\n m/s \n70\nMpa']/1000  #vp km/h
vs1=arqdt['VS1\n m/s \n70\nMpa']/1000  #vs1 km/h
vs2=arqdt['VS2\n m/s \n70\nMpa']/1000  #vs2 km/h
phi=arqdt['Porosity  (%) ']  #porosidade
```

```python
#Gráfico 1

#gerar plot vp-phi usando a função
plot_figure(phi,vp,'Vp-\phi','Vp (km/s)',0,60,1,7)  
```

```python
#Gráfico 2

#gerar plot vs1-phi usando a função
plot_figure(phi,vs1,'Vs1-\phi','Vs1 (km/s)',0,60,0,5) 
```

```python
#Gráfico 3

#gerar plot vs2-phi usando a função
plot_figure(phi,vs2, 'Vs2-\phi','Vs2 (km/s)',0,60,0,5)  #gerar plot vs2-phi
```
