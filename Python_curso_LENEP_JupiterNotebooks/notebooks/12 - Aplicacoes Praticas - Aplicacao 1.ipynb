## Aplicações Práticas - Aplicação 1

 
  
Nesta aplicação prática será lido um arquivo de precipitação de chuvas no Aeroporto Internacional de Portland (dados_precipitacao.dat) que é um arquivo csv separado por tabulações.  
A partir desses dados, serão exibidos alguns gráficos.

```python
#Primeira parte - importar as bibliotecas necessárias
#Neste exercício serão importadas as bibliotecas do pandas para manipulação do arquivo e 
#do matplotlib para plotagem dos gráficos

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Abrindo o arquivo dat que na verdade é um csv separado por tab (\t)
#Além disso, o arquivo não tem header, e na abertura já define os seus cabeçalhos
#A primeira coluna possui os meses e na segunda coluna a precipitação

arq = pd.read_csv("dados_precipitacao.dat", sep = "\t", header=None, names=['month','precip'])
print(arq.head)
```

```python
#Calculando a média de precipitação
#Formatando a media com 2 casas decimais na saída

print("Média Anual: %0.2f" % arq['precip'].mean())
```

```python
#Plotagem do primeiro gráfico: Precipitação vs Mês

plt.figure()

#plota precip vs month
plt.plot(arq['month'], arq['precip'], 'o') 
 
#Coloca título no gráfico
plt.title("Média de precipitação mensal no Aeroporto Internacional Portland")

#Mostra o texto com a média geral no gráfico na posição x=1, y=6
plt.text(1,6,"Média Anual: %0.2f" % arq['precip'].mean())  

#Coloca uma label no eixo de X
plt.xlabel('Mês')

#Coloca uma label no eixo de y
plt.ylabel('Precipitação média (polegadas)') 
```

```python
#Plotagem do segundo gráfico: Barras de Precipitação vs Mês

plt.figure()

#Gráfico de barras precip vs month
plt.bar(arq['month'], arq['precip']) 

#Mostra o texto com a média geral no gráfico na posição x=1, y=6
plt.text(1,6,"Média Anual: %0.2f" % arq['precip'].mean())  

#Coloca título no gráfico
plt.title("Média de precipitação mensal no Aeroporto Internacional Portland")  

#Mostra o texto com a média geral no gráfico na posição x=1, y=6
plt.text(1,6,"Média Anual: %0.2f" % arq['precip'].mean())  

#Coloca uma label no eixo de X
plt.xlabel('Mês') 

#Coloca uma label no eixo de y
plt.ylabel('Precipitação média (polegadas)') 
```

```python
#Plotagem do terceiro gráfico: Escada de Precipitação vs Mês


plt.figure()

#Gráfico tipo escada precip vs month
plt.step(arq['month'], arq['precip'])

#Mostra o texto com a média geral no gráfico na posição x=1, y=6
plt.text(1,6,"Média Anual: %0.2f" % arq['precip'].mean())  

#Coloca título no gráfico
plt.title("Média de precipitação mensal no Aeroporto Internacional Portland")  

#Mostra o texto com a média geral no gráfico na posição x=1, y=6
plt.text(1,6,"Média Anual: %0.2f" % arq['precip'].mean())  

#Coloca uma label no eixo de X
plt.xlabel('Mês') 

#Coloca uma label no eixo de y
plt.ylabel('Precipitação média (polegadas)') 
```

```python
#Plotagem do quarto gráfico: Gráfico de Crescimento


plt.figure()

#Gráfico de crescimento
plt.stem(arq['month'], arq['precip']) 

#Mostra o texto com a média geral no gráfico na posição x=1, y=6
plt.text(1,6,"Média Anual: %0.2f" % arq['precip'].mean())  

#Coloca título no gráfico
plt.title("Média de precipitação mensal no Aeroporto Internacional Portland")  

#Mostra o texto com a média geral no gráfico na posição x=1, y=6
plt.text(1,6,"Média Anual: %0.2f" % arq['precip'].mean())  

#Coloca uma label no eixo de X
plt.xlabel('Mês') 

#Coloca uma label no eixo de y
plt.ylabel('Precipitação média (polegadas)') 
```
