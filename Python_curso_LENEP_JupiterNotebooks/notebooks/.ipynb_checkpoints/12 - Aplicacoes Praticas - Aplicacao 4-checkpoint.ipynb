## Aplicações Práticas - Aplicação 3

 
  
Nesta aplicação prática serão lidos dois arquivos de dados de poços (poco1_adaptado.xls e poco1_adaptado_cego.xls) que são arquivos xls (Excel).  
A partir desses dados, serão exibidos alguns gráficos, com alguns dados como estão na planilha, outros com os dados alterados ou com variáveis geradas a partir desses dados

```python
#Primeira parte - importar as bibliotecas necessárias
#Neste exercício serão importadas as bibliotecas do pandas para manipulação do arquivo e 
#do matplotlib para plotagem dos gráficos

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Abre o primeiro arquivo xls
arqdt = pd.read_excel("poco1_adaptado.xls")
print(arqdt.head(8))
```

```python
#Na ordem da planilha... Capturando dados de cada coluna pelos seus nomes...
#Observem que em alguns cabeçalhos foram usandos caracteres como \n e espaços pois é assim que a planilha foi 
#preenchida. Para facilitar, usou-se a impressão das colunas para poder usar o nome do cabeçalho com mais precisão

#Obs.: \n representa nova linha
print(arqdt.columns)


prof=arqdt['Depth\n (m) '] #profundidade
lito=arqdt['Lithology'] #litologia
por=arqdt['Porosity  (%) ']  #porosidade
rhob=arqdt['Dry bulk density  g/cm3 ']  #densidade
vp=arqdt['VP\n m/s \n70\nMpa']/1000  #vp km/h
vs1=arqdt['VS1\n m/s \n70\nMpa']/1000  #vs1 km/h
vs2=arqdt['VS2\n m/s \n70\nMpa']/1000  #vs2 km/h
k1=arqdt['K1\n GPa \n70\nMpa'] #Módulos de incompressibilidade
k2=arqdt['K2\n GPa \n70\nMpa']
g1=arqdt['m1\n GPa \n70\nMpa'] #Módulos de cisalhamento
g2=arqdt['m2\n GPa \n70\nMpa']
gs=arqdt['Grain size']
micf=arqdt['Microporous carbonate fraction']
calcite=arqdt['Calcite']
dolomite=arqdt['Dolomite']
quartz=arqdt['Quartz']
orthoclase=arqdt['Orthoclase']
albite=arqdt['Albite']
clay=arqdt['Clays']
pyrite=arqdt['Pyrite']
inckt=arqdt['α KT (inc poros)']
mine=arqdt['Mineralogia']
text=arqdt['Textura']
```

```python
#Calculando propriedade elásticas e densidade mineral

ko=(71*calcite+76.4*dolomite+37*quartz+37.5*orthoclase+75.6*albite+147.4*pyrite)/100
go=(31*calcite+49.7*dolomite+45*quartz+15*orthoclase+25.6*albite+132.5*pyrite)/100
dos=(2.71*calcite+2.87*dolomite+2.65*quartz+2.62*orthoclase+2.63*albite+4.93*pyrite)/100
```

```python
#Carregar o segundo arquivo com dados acessórios acessórios
arqdt = pd.read_excel("poco1_adaptado_cego.xls")
```

```python
#Definindo as demais variáveis a partir do segundo arquivo
#Usando a mesma forma: imprimindo os cabeçalhos para mais fácil detectar como eles estão no arquivo
print(arqdt.columns)

prof2=arqdt['Depth\n (m) ']
lito2=arqdt['Lithology']
por2=arqdt['Porosity  (%) ']
```

```python
#Primeiro Gráfico
plt.figure()

#Um plot com todos os dados abaixo
plt.plot(calcite,prof,quartz,prof,dolomite,prof,orthoclase,prof,albite,prof,clay,prof,pyrite,prof,linewidth=2)

#Mostra a grif
plt.grid()

#Inverte o eixo (imprime de 'cabeça para baixo')
plt.gca().invert_yaxis()

#Título do gráfico
plt.title('Mineral')

#Eixo de X
plt.xlabel('(%)')

#Eixo de Y
plt.ylabel('Depth(m)')

#Legendas para todos os dados
plt.legend(['Calcite','Quartz','Dolomite','Orthoclase','Albite','Clay','Pyrite'],fontsize=14)

#Determina os limites dos eixos:
#x - de 0 a 100
#y - -5 a 155
plt.axis([0,100,-5,155])
```

```python
#Segundo gráfico

plt.figure

#Gráfico de linha Porosidade Vs Profundidade
plt.plot(por,prof,'--',linewidth=2)

#Gráfico de dispersão
plt.scatter(por2,prof2,150,lito2,'d','jet',linewidth=2)

#Grade e Caixa
plt.grid()
plt.box()

#Inverte o eixo de y
plt.gca().invert_yaxis()

#Imprime uma barra de cores
plt.colorbar()

#Imprime o título
plt.title('Porosity(\phi)')

#Label do eixo de X
plt.xlabel('(%)')

#Definindo os dados dos eixos
#x - de -1 a maior porosidade+1
#y - de -5 a 155
plt.axis([-1, por.max()+1,-5,155])
plt.draw
```

```python
#Terceiro gráfico

plt.figure

#Plotando os dados
plt.plot(dos,prof,'b-',rhob,prof,'k-',linewidth=2)

#Criando uma grade e caixa
plt.grid()
plt.box()

#Inverte o eixo de y
plt.gca().invert_yaxis()

#Exibe a legenda
plt.legend(['Mineral','Dry'])

#Exibe o título
plt.title('Density')

#Exibe um título para o eixo de X
plt.xlabel('(g/cc)')

#Define os valores dos eixos
#x - menor valor de densidade -0.05 até o maior valor de dos
plt.axis([ min(rhob)-0.05, max(dos)+0.05, -5, 155]);
plt.draw
```

```python
#Quarto gráfico

plt.figure

#Plotando os dados
plt.plot(ko,prof,'b-',(k1+k2)/2,prof,'k-',linewidth=2)

#Exibe grade e caixa
plt.grid()
plt.box()

#Inverte o eixo de y
plt.gca().invert_yaxis()

#Exibe a legenda
plt.legend(['Mineral','Dry'])

#Exibe o título
plt.title('Bulk modulus')

#Exibe o título do eixo de X
plt.xlabel('(GPa)')

#Determina os limites dos eixos
plt.axis([5, max(ko)+10, -5, 155]);


plt.draw
```

```python
#Quinto gráfico

plt.figure

#Plotando os dados
plt.plot(go,prof,'b-',(g1+g2)/2,prof,'k-',linewidth=2)

#Exibindo as grades e caixas
plt.grid()
plt.box()

#Inverte o eixo de y
plt.gca().invert_yaxis()

#Exibe a legenda
plt.legend(['Mineral','Dry'])

#Exibe o título
plt.title('Shear modulus')

#Exibe o título do eixo de X
plt.xlabel('(GPa)')

#Determina os limites dos eixos
plt.axis([0, max(go)+5, -5, 155]);

plt.draw
```

```python
#Plotando o gráfico
#Na verdade serão 5 gráficos juntos num único gráfico. 
#Por isso o uso dos subplots em um único plot
#plt.subplot(1on='true',5,x) - 1 linha, 5 colunas e o terceiro é o identificado do gráfico que vai do 1 ao 5


plt.figure()

#Primeiro subplot
plt.subplot(1,5,1)

plt.plot(calcite,prof,quartz,prof,dolomite,prof,orthoclase,prof,albite,prof,clay,prof,pyrite,prof,linewidth=2)
plt.grid()
plt.gca().invert_yaxis()
plt.title('Mineral')
plt.xlabel('(%)')
plt.ylabel('Depth(m)')
plt.legend(['Calcite','Quartz','Dolomite','Orthoclase','Albite','Clay','Pyrite'],fontsize=14)
plt.axis([0,100,-5,155])


#Segundo subplot
plt.subplot(1,5,2)

plt.plot(por,prof,'--',linewidth=2)
plt.scatter(por2,prof2,150,lito2,'d','jet',linewidth=2)
plt.grid()
plt.box()
plt.gca().invert_yaxis()
plt.colorbar()
plt.title('Porosity(\phi)')
plt.xlabel('(%)')
plt.axis([-1, por.max()+1,-5,155])
plt.draw

#Terceiro subplot
plt.subplot(1,5,3)
plt.plot(dos,prof,'b-',rhob,prof,'k-',linewidth=2)
plt.grid()
plt.box()
plt.gca().invert_yaxis()
plt.legend(['Mineral','Dry'])
plt.title('Density')
plt.xlabel('(g/cc)')
plt.axis([ min(rhob)-0.05, max(dos)+0.05, -5, 155]);
plt.draw

#Quarto subplot
plt.subplot(1,5,4)
plt.plot(ko,prof,'b-',(k1+k2)/2,prof,'k-',linewidth=2)
plt.grid()
plt.box()
plt.gca().invert_yaxis()
plt.legend(['Mineral','Dry'])
plt.title('Bulk modulus')
plt.xlabel('(GPa)')
plt.axis([5, max(ko)+10, -5, 155]);
plt.draw

#Quinto subplot
plt.subplot(1,5,5)
plt.plot(go,prof,'b-',(g1+g2)/2,prof,'k-',linewidth=2)
plt.grid()
plt.box()
plt.gca().invert_yaxis()
plt.legend(['Mineral','Dry'])
plt.title('Shear modulus')
plt.xlabel('(GPa)')
plt.axis([0, max(go)+5, -5, 155]);
plt.draw
```
