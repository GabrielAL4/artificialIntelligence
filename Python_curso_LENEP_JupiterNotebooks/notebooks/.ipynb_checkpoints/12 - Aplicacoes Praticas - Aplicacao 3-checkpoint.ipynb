## Aplicações Práticas - Aplicação 2

 
  
Nesta aplicação prática será lido um arquivo de precipitação de chuvas no Aeroporto Internacional de Portland (dados_precipitacao.xls) que é um arquivo xls (Excel).  
A partir desses dados, serão exibidos alguns gráficos.

Uma característica é que esse arquivo não inicia os dados na primeira linha

```python
#Primeira parte - importar as bibliotecas necessárias
#Neste exercício serão importadas as bibliotecas do pandas para manipulação do arquivo e 
#do matplotlib para plotagem dos gráficos

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Abre o arquivo xls
arq = pd.read_excel("dados_temperatura.xls")
print(arq.head(8))
```

```python
#Observando o arquivo, ele só inicia as linhas de dados a partir da 5a linha e nelas perde a referência 
#dos cabeçalhos. Assim, pode-se definir manualmente a partir de qual linha se inicia

arq = pd.read_excel("dados_temperatura.xls",  header=5)
print(arq)
```

```python
#Plotando o primeiro gráfico
plt.figure()

#Temperaturas altas, médias e baixas pelos meses
plt.plot(arq['Mês'], arq['Alta'], 'ro', arq['Mês'], arq['Baixa'], 'c+', arq['Mês'], arq['Média'], 'g-') 

#Coloca título no gráfico
plt.title("Balanço de temperatura mensal no aeroporto internacional Portland(1)")  

#Cria uma legenda com os dados gerados no gráfico
plt.legend(['Alta', 'Baixa', 'Mèdia'])  

#Coloca uma label no eixo de X
plt.xlabel('Mês') 

#Coloca uma label no eixo de y
plt.ylabel('Temperatura (F)') 
```

```python
#Plotando o segundo gráfico
plt.figure()

#Gráfico de barras Mês Vs Média
plt.bar(arq['Mês'], arq['Média'])

#Plota agora várias vezes no mesmo gráfico
#1) Mês Vs Temperatura Alta com pontos vermelhor (ro)
#2) Mês vs Temperatura Baixa com cruzes (c+)
#3) Mês vs Temperatura Média com Linha (LineWidth=1.2)
plt.plot(arq['Mês'], arq['Alta'], 'ro', arq['Mês'], arq['Baixa'], 'c+', arq['Mês'], arq['Média'], LineWidth=1.2, MarkerSize=10)

#Cria uma caixa e grades no gráfico
plt.grid()
plt.box()

#Cria uma legenda com os dados gerados no gráfico
plt.legend(['Alta', 'Baixa', 'Mèdia', 'Bar-Média'])

#Coloca uma label no eixo de X
plt.xlabel('Mês') 

#Coloca uma label no eixo de y
plt.ylabel('Temperatura (F)') 

#Coloca título no gráfico
plt.title("Balanço de temperatura mensal no aeroporto internacional Portland(2)")  
```
