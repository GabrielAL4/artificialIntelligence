# A Matriz de Correlação, baseada no coeficiente de Pearson, avalia relações lineares entre pares de variáveis independentes. Valores próximos a 1 indicam correlações positivas fortes, próximos a -1 indicam correlações negativas fortes, e valores próximos a 0 indicam correlações fracas. Essa ferramenta pode ser aplicada com Python e matplotlib para visualização prática.

# Suponhamos que você tenha um DataFrame com seus dados
# Vamos criar dados fictícios para ilustração.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.DataFrame({
    'Feature1': np.random.rand(100),
    'Feature2': np.random.rand(100),
    'Feature3': np.random.rand(100)
})
```

# Aqui vamos calcular a matriz de correlação de Pearson e Plotar um gráfico de calor com base na mesma

```python
correlation_matrix = data.corr()

plt.figure(figsize=(8, 6))
plt.matshow(correlation_matrix, cmap='coolwarm', fignum=1)
plt.colorbar(label='Correlation Coefficient')
plt.xticks(range(len(correlation_matrix.columns), correlation_matrix.columns, rotation=90))
plt.yticks(range(len(correlation_matrix.columns), correlation_matrix.columns))
plt.title('Matriz de Correlação de Pearson')
plt.show()
```
