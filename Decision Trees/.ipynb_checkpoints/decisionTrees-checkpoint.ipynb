# Árvores de Decisão
### As árvores de decisão tratam-se de um modelo de aprendizado de máquina, este modelo toma decisões baseado em condições lógicas. De certa forma acaba sendo mais fácil pensar de maneira em que são feitas perguntas que resultam em "sim/não" e estas perguntas resultam em decisões diferentes.

### Na imagem abaixo é possível visualizar, entendendo que cada "nó" tratasse de uma pergunta ou uma condição e cada "galho" por sua vez trata-se de uma possível resposta para este "nó".



# Classes de pétalas de Iris

### Neste contexto a árvore se volta para realizar perguntas sobre as características das flores(comprimento e largura da pétala por exemplo)e só então, baseando-se em suas respostas, ela realiza a classificação da cuja flor em uma categoria específica(setosa, versicolor ou virginica).

### Abaixo temos um exemplo ilustrativo de como funcionaria esta classificação:



```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Carregar produto de dados 

iris = load_iris()
x = iris.data
y = iris.target

# Iniciando o treinamento com os conjuntos de dados divididos

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 

# CLassificador da árvore de decisão instanciado
classifier = DecisionTreeClassifier()

# Ajustando o modelo com o classificador treinando

classifier.fit(x_train, y_train)

# Visualização da árvore de decisão

plt.figure(figsize=(12, 8))
plot_tree(classifier, filled=True, feature_names=iris['feature_names'], class_names=iris.target_names.tolist(), rounded=True)
plt.show()
```

# Previsões e estimativas 
### Após termos uma árvore de decisões treinada é possível utiliza-la para realizar previsões e estimativas a cerca de resultados futuros. 

### Neste caso no código abaixo iremos completar o código anterior adícionando técnicas de previsão.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Carregar produto de dados 

iris = load_iris()
x = iris.data
y = iris.target

# Iniciando o treinamento com os conjuntos de dados divididos

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 

# CLassificador da árvore de decisão instanciado
classifier = DecisionTreeClassifier()

# Ajustando o modelo com o classificador treinando

classifier.fit(x_train, y_train)

# Visualização da árvore de decisão

plt.figure(figsize=(12, 8))
plot_tree(classifier, filled=True, feature_names=iris['feature_names'], class_names=iris.target_names.tolist(), rounded=True)
plt.show()

# Aqui iremos ealizar previsões acerca de nosso modelo treinado

y_pred = classifier.predict(x_test)
print("Previsões:", y_pred)

# Dado as previsões também conseguimos gerar estimativas prováveis a cerca de nosso modelo

predict_prob = classifier.predict_proba(x_test)

print("Estimativas de Probabilidade:")
for i, class_name in enumerate(iris.target_names):
    print(f"{class_name}: {predict_prob[0][i]:.4f}")
    
    
```
