# rvores de Deciso
### As rvores de deciso tratam-se de um modelo de aprendizado de mquina, este modelo toma decises baseado em condies lgicas. De certa forma acaba sendo mais fcil pensar de maneira em que so feitas perguntas que resultam em "sim/no" e estas perguntas resultam em decises diferentes.

### Na imagem abaixo  possvel visualizar, entendendo que cada "n" tratasse de uma pergunta ou uma condio e cada "galho" por sua vez trata-se de uma possvel resposta para este "n".



# Classes de ptalas de Iris

### Neste contexto a rvore se volta para realizar perguntas sobre as caractersticas das flores(comprimento e largura da ptala por exemplo)e s ento, baseando-se em suas respostas, ela realiza a classificao da cuja flor em uma categoria especfica(setosa, versicolor ou virginica).

### Abaixo temos um exemplo ilustrativo de como funcionaria esta classificao:



```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Carregar produto de dados 

iris = load_iris()
x = iris.data
y = iris.target

# Iniciando o treinamento com os conjuntos de dados divididos

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 

# CLassificador da rvore de deciso instanciado
classifier = DecisionTreeClassifier()

# Ajustando o modelo com o classificador treinando

classifier.fit(x_train, y_train)

# Visualizao da rvore de deciso

plt.figure(figsize=(12, 8))
plot_tree(classifier, filled=True, feature_names=iris['feature_names'], class_names=iris.target_names.tolist(), rounded=True)
plt.show()
```

# Previses e estimativas 
### Aps termos uma rvore de decises treinada  possvel utiliza-la para realizar previses e estimativas a cerca de resultados futuros. 

### Neste caso no cdigo abaixo iremos completar o cdigo anterior adcionando tcnicas de previso.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Carregar produto de dados 

iris = load_iris()
x = iris.data
y = iris.target

# Iniciando o treinamento com os conjuntos de dados divididos

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 

# CLassificador da rvore de deciso instanciado
classifier = DecisionTreeClassifier()

# Ajustando o modelo com o classificador treinando

classifier.fit(x_train, y_train)

# Visualizao da rvore de deciso

plt.figure(figsize=(12, 8))
plot_tree(classifier, filled=True, feature_names=iris['feature_names'], class_names=iris.target_names.tolist(), rounded=True)
plt.show()

# Aqui iremos ealizar previses acerca de nosso modelo treinado

y_pred = classifier.predict(x_test)
print("Previses:", y_pred)

# Dado as previses tambm conseguimos gerar estimativas provveis a cerca de nosso modelo

predict_prob = classifier.predict_proba(x_test)

print("Estimativas de Probabilidade:")
for i, class_name in enumerate(iris.target_names):
    print(f"{class_name}: {predict_prob[0][i]:.4f}")
    
    
```
